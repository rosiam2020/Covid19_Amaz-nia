```python
assimetria = aval1['calorias'].skew()
print('Assimetria(Skewness calorias) ', assimetria)
assimetria = aval1['Proteinas'].skew()
print('Assimetria(Skewness proteina) ', assimetria)
assimetria = aval1['Gorduras'].skew()
print('Assimetria(Skewness Gorduras) ', assimetria)
assimetria = aval1['Sódio'].skew()
print('Assimetria(Skewness Sódio) ', assimetria)
assimetria = aval1['Fibra'].skew()
print('Assimetria(Skewness Fibra) ', assimetria)
assimetria = aval1['Açucares'].skew()
print('Assimetria(Skewness açucares) ', assimetria)
assimetria = aval1['Potássio'].skew()
print('Assimetria(Skewness Potássios) ', assimetria)
assimetria = aval1['Vitaminas'].skew()
print('Assimetria(Skewness Vitaminas) ', assimetria)
```

Respostas:
Variáveis simétricas são as que possuem assimetria entre -1 e 1, são elas: Calorias, Proteínas, Sódios e Açucares.
As variáveis que possuem assimetria positiva são: Gorduras, Fibras, Potássio e vitaminas. Nenhuma vaiável apresentou assimetria negativa.

```python

curtose = aval1['calorias'].kurtosis()
print('Curtose calorias) ', curtose)
curtose = aval1['Proteinas'].kurtosis()
print('curtose proteina) ', curtose)
curtose = aval1['Gorduras'].kurtosis()
print('curtose Gorduras) ', curtose)
curtose = aval1['Sódio'].kurtosis()
print('curtose Sódio) ', curtose)
curtose = aval1['Fibra'].kurtosis()
print('curtose Fibra) ', curtose)
curtose = aval1['Açucares'].kurtosis()
print('curtose açucares) ', curtose)
curtose = aval1['Potássio'].kurtosis()
print('curtose Potássios) ', curtose)
curtose = aval1['Vitaminas'].kurtosis()
print('curtose Vitaminas) ', curtose)
```

Variáveis mesocurticas:  sódio
Variáveis Platicurtica : (curtose < 0) - açucares.
são séries com distribuição mais achatadas quando comparadas às mesocurticas

Variáveis Leptocúrtica: (curtose > 0) - calorias, porteína, gorduras, fibra, portássio e vitaminas
são séries com distribuição mais afinadas e alongadas quando comparadas às mesocurticas.
