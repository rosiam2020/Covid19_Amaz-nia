```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import date
import warnings
warnings.filterwarnings ('ignore')
```

```python
casos_14 = pd.read_excel ('modelos_casos_14_teste.xlsx')
casos_14.head ()
```

```python
df = casos_14
df['data'] = pd.to_datetime(df['Data']).dt.date
df.head()
```

```python
df['temp_max']=df['Temp. Max. (C)']
df['temp_min']=df['Temp. Min. (C)']
df['temp']=df['Temp. Ins. (C)']
df['umi']=df['Umi. Ins. (%)']
df['umi_max']=df['Umi. Max. (%)']
df['umi_min']=df['Umi. Min. (%)']
df['casos']=df['mediacasos_14']
df['rest']=df['restricao']
df['resp']=df['governo']
df['apoio']=df['apoio']
df['sau']=df['saude']
df['knn']=df['kNN']
df['ran']=df['Random Forest']
df['regr']=df['Linear Regression']
df['grad']=df['Gradient Boosting']
df['ada']=df['AdaBoost']

```

```python
x=df['data']
y2=df['temp']
y3=df['umi']
y4=df['knn']
y5=df['ran']
y1=df['regr']
y6=df['grad']
y7=df['ada']
y8=df['casos']


```

```python
plt.rcParams.update({'font.size':16})
```

```python
print(df)
column_1 = df["casos"]
column_2 = df["regr"]
correlation = column_1. corr(column_2) 
print(correlation)
```

```python
x=df['data']
y4=df['regr']
y5=df['casos']
fig, ax1 = plt.subplots(figsize=(12,6))
ax1.set_xlabel('Data')
ax1.set_ylabel('Média Móvel de Casos de Covid (14 dias) ', color='red')
ax1.plot(x, y5, color='red', marker = 'o', markersize = 3, linestyle= '--')

ax2 = ax1.twinx()  
ax2.set_ylabel('Previsão com Modelo de Regressão Linear', color='green')  

ax2.plot(x, y4, color = 'green',  marker = 's', markersize = 3, linestyle= '-.')
plt.title("Média Móvel de Casos de Covid (14 dias) x Previsão com Modelo de Regressão Linear - corr 0.864" , 
          fontdict={'family': 'serif', 
                    'color' : 'Black',
                    'weight': 'bold',
                    'size': 16},
          loc='left')

plt.show()
```
